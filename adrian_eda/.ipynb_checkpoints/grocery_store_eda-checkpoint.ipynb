A notebook for exploration of a data set being potentially mined through association analysis.

```python
#Imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import squarify
import seaborn as sns

from apyori import apriori

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
#Set options
pd.options.display.max_rows = 10
pd.options.display.float_format = '{:,.2f}'.format #suppress scientific notation
plt.rcParams['figure.figsize'] = (15, 15)
```

```python
#Load data
grocery_store = pd.read_csv("../Market_Basket_Optimisation.csv", header=None)
```


Data Visualizations


```python
#Most Frequent Items
plt.rcParams['figure.figsize'] = (15,15)
color = plt.cm.copper(np.linspace(0, 1, 50))
grocery_store[0].value_counts().head(20).plot.bar( color = color)
plt.title("Top 20 Most Frequent Items")
plt.ylabel('Counts')
plt.xlabel('Items')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
#Word cloud
plt.rcParams['figure.figsize'] = (15,15)
wordcloud = WordCloud(background_color = 'white',
               width = 1200,  height = 1200,
               max_words = 121).generate(str(grocery_store[0]))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Most Popular Items in Shopping Basket", fontsize = 20)
plt.show()
```

```python
#Tree map
y = grocery_store[0].value_counts().head(50).to_frame()
y.index

plt.rcParams['figure.figsize'] = (15,15)
color = plt.cm.cool(np.linspace(0, 1, 50))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color)
plt.title('Tree Map for Popular Items')
plt.axis('off')
plt.show()
```

```python
#Derive rows from set by index
grocery_rows = []
for m in range(0, len(grocery_store)):
    grocery_rows.append([str(grocery_store.values[m,n]) for n in range(0,20)])   
grocery_rows[0]
```
